### OneHot coding 

```python
df['Qty'].values
```

```python
df['Gender'] = np.where(df['Gender'].isin(['M', 'Men']), 1, 0)


```

```python
word_to_num = {
    'one': 1,
    'two': 2,
    'three': 3,
    'four': 4,
    'five': 5,
    'six': 6,
    'seven': 7,
    'eight': 8,
    'nine': 9,
    'ten': 10
}
def convert_qty(value):
    try:
        # If it's already an integer or float, return as int
        return int(value)
    except:
        # If it's a string like "one", convert using the dictionary
        return word_to_num.get(str(value).strip().lower(), None)  # returns None if not found

df['Qty'] = df['Qty'].apply(convert_qty)
```

```python
df['Qty'].nunique()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['Channel'] = label_encoder.fit_transform(df['Channel'])  # e.g., 'Flipkart', 'Amazon', etc.
```

### Change order status

```python
df['Status'] = np.where(df['Status'].isin(['Delivered']), 1, 0)
```

```python
df['B2B'] = np.where(df['B2B'].isin(['True']), 1, 0)
```

```python
df.head()
```

```python
y = df['Status']
x = df[['Gender', 'Age', 'Channel', 'SKU', 'Category', 'Size', 'Qty', 'Amount',
     'ship-city', 'ship-state', 'B2B', 'order_day', 'order_month',
       'order_weekday']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train
```

```python
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

```

```python
model = CatBoostClassifier()
cat_features = ['SKU','Category','Size','ship-city','ship-state']

model.fit(x_train, y_train, cat_features)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
print(accuracy_score(y_test,y_pred))
```
